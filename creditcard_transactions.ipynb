```python
df['birthdate'] = pd.to_datetime(
    df['birthdate'], format='%d-%m-%Y', errors='coerce'
)

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df
```

```python
df ['date'] = df_raw['Date']
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce', dayfirst=True)
df
```

```python
df.dropna(subset=['birthdate', 'date'], inplace=True)
df
```

```python
df['transaction_amount'] = pd.to_numeric(
    df['transaction_amount'], errors='coerce'
)

df = df[df['transaction_amount'] > 0]
```

```python
df
```

```python
df.duplicated().sum()
```

```python
text_cols = ['name', 'surname', 'merchant_name', 'category']

for col in text_cols:
    df[col] = df[col].str.strip().str.title()
df
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day_name'] = df['date'].dt.day_name()
df
```

```python
df['age'] = (df['date'] - df['birthdate']).dt.days // 365
df
```

```python
df = df[(df['age'] >= 18) & (df['age'] <= 100)]
df
```
